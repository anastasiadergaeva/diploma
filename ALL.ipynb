```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nmax = 201
```

```python
OscilloscopeALL = pd.read_csv('ALL_PERIOD.TXT', header=None)
OscilloscopeALL
```

```python
OscilloscopeALL = OscilloscopeALL.transpose()
OSCData_ALL = OscilloscopeALL[0:Nmax]
OSCData_ALL.values
```

```python
y = OSCData_ALL.values
x = range(len(OSCData_ALL))
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Полная нагрузка')
plt.title('ОСЦИЛЛОГРАММА')
plt.xlabel('Номер точки измерения')
plt.ylabel('Сигнал осциллографа')
plt.grid()
plt.legend()
plt.show()
```
