```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nmax = 200
```

```python
ELECTROLamp = pd.read_csv('ELECTRO_PERIOD.TXT', header=None)
ELECTROLamp
```

```python
ELECTROLamp = ELECTROLamp.transpose()
ELECTROLamp_data = ELECTROLamp[0:Nmax]
ELECTROLamp_data.values
```

```python
y = ELECTROLamp_data.values
x = range(len(ELECTROLamp_data))
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Лампа накаливания')
plt.title('ОСЦИЛЛОГРАММА')
plt.xlabel('Номер точки измерения')
plt.ylabel('Сигнал осциллографа')
plt.grid()
plt.legend()
plt.show()
```
